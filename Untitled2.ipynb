```python
import numpy as np
from sklearn.decomposition import PCA
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
```

```python
a= read('scene01.wav')
print(a[1])
list=np.array(a[1])
```

```python
pca=PCA(n_components=4)
pca.fit(list)
```

```python
 s= pca.transform(list)
```

```python
print(s)
```

```python
s.shape
```

```python
plt.figure()
model=[list,s]
color=['red','yellow']
for ii, mo in enumerate(zip(model,color),1):
    plt.subplot(2,1,ii)
    for sig,colo in zip(mo,color):
        plt.plot(sig,color=colo)
        
#plt.subplots_adjust(0.09,0.04)
plt.show()
```

```python
from sklearn.decomposition import FastICA
```

```python
ica = FastICA(n_components=4)
```

```python
s=ica.fit_transform(s)
a=ica.mixing_
print(s)

```

```python
plt.figure()
model=[list,s,]
colors=['red','yellow']
for ii, mo in enumerate(zip(model,colors),1):
    plt.subplot(2,1,ii)
    for sig,colo in zip(mo,color):
        plt.plot(sig,color=colo)
        
#plt.subplots_adjust(0.09,0.04)
plt.show()

```

```python
plt.figure()
model=[a,a[0],a[1],a[2],a[3]]
colors=['red','yellow','steelblue','orange','yellow']
for ii, mo in enumerate(zip(model,colors),1):
    plt.subplot(5,1,ii)
    for sig,colo in zip(mo,color):
        plt.plot(sig,color=colo)
        
#plt.subplots_adjust(0.09,0.04)
plt.show()
```
